__1.__ Исследовать на линейную зависимость:
$$f_{1}(x)=e^{x}, f_{2}(x)=1, f_{3}(x)=x+1, f_{4}(x)=x-e^{x}.$$

Решение:

$$f_{4}(x)=f_{3}(x)-f_{2}(x)-f_{1}(x)$$

Тогда $$f_{4}(x)$$ - линейная комбинация векторов:

$$f_{1}(x),f_{2}(x),f_{3}(x)$$

Следовательно эти вектора линейно зависимы.

__2.__ Исследовать на линейную зависимость:

$$f_{1}(x)=2, f_{2}(x)=x, f_{3}(x)=x^{2}, f_{4}(x)=(x+1)^{2}$$

Решение:
$$f_{4}(x)=f_{3}(x)+2*f_{2}(x)+1/2*f_{1}(x)$$

Тогда $$f_{4}(x)$$ - линейная комбинация векторов:  
$$f_{1}(x),f_{2}(x),f_{3}(x)$$

Следовательно эти вектора линейно зависимы.

$$x=b_{2}+3b_{3}+0.5b_{1}$$


то есть координатами вектора $$x$$ в стандартном базисе являются 1, 3, 0.5.
